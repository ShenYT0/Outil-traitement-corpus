```python
from datasets import load_dataset
dataset = load_dataset("json", data_files="../../data/raw/cocktailclub_recipes.json", split="train")
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def static_token_count(line):
    text_ingredients = ""
    for ingredient in line["ingredients"]:
        text_ingredients += ingredient + " "
    line["nb_token_ingredients"] = len(nlp(text_ingredients))
    line["text_length_ingredients"] = len(text_ingredients)
    line["text_ingredients"] = text_ingredients

    text_misc = ""
    for misc in line["misc"]:
        text_misc += misc + " "
    line["nb_token_misc"] = len(nlp(text_misc))
    line["text_length_misc"] = len(text_misc)
    line["text_misc"] = text_misc

    text_directions = ""
    for direction in line["directions"]:
        text_directions += direction + " "
    line["nb_token_directions"] = len(nlp(text_directions))
    line["text_length_directions"] = len(text_directions)
    line["text_directions"] = text_directions

    text_all = text_ingredients + text_misc + text_directions
    line["nb_token_all"] = len(nlp(text_all))
    line["text_length_all"] = len(text_all)
    line["text_all"] = text_all

    return line

dataset = dataset.map(static_token_count)

dataset[0]
```

```python
# save the dataset as csv
dataset_df = pd.DataFrame(dataset)
dataset_df.to_csv("../../data/raw/analyzed_cocktailclub_recipes.csv", index=False)
```

```python
import sqlite3
con = sqlite3.connect("../../data/raw/cocktailclub_recipes.db")

df = pd.read_csv("../../data/raw/analyzed_cocktailclub_recipes.csv")

df.to_sql(con = con, name = 'cocktailclub_recipes', if_exists='replace', index=False)
```

```python
# SQL query
query = "SELECT * FROM cocktailclub_recipes"
db = pd.read_sql_query(query, con)
db
```

```python
# P-value
import numpy as np
total_length_ingredients = db["text_length_ingredients"]
total_length_misc = db["text_length_misc"]
np.corrcoef(total_length_ingredients, total_length_misc)
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
train, test = train_test_split(db, test_size=0.2)
```

```python
# save train and test as csv
train.to_csv("../../data/clean/train_cocktailclub_recipes.csv", index=False)
test.to_csv("../../data/clean/test_cocktailclub_recipes.csv", index=False)
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(db["text_all"])
cosine_sim = cosine_similarity(X, X)

# Recommender system
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = db[db["title"] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    recipe_indices = [i[0] for i in sim_scores]
    return db["title"].iloc[recipe_indices]

get_recommendations("Mai Tai")

# Save the model
import joblib
joblib.dump(vectorizer, "../../src/model1/vectorizer.pkl")
```

```python
# test the model
vectorizer = joblib.load("../../src/model1/vectorizer.pkl")
X = vectorizer.transform(db["text_all"])
cosine_sim = cosine_similarity(X, X)
get_recommendations("Mai Tai", cosine_sim)
```
