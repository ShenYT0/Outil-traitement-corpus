```python
import pandas as pd
# load the database from the csv file
df = pd.read_csv("../../data/raw/analyzed_cocktailclub_recipes.csv")
```

```python
# chart text length ingredients
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="text_length_ingredients", bins=50)
plt.title("Text length ingredients")
plt.show()

# chart text length misc

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="text_length_misc", bins=50)
plt.title("Text length misc")
plt.show()

# chart text length directions

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="text_length_directions", bins=50)
plt.title("Text length directions")
plt.show()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

# chart top 10 word frequency ingredients
from collections import Counter
import matplotlib.pyplot as plt

text_ingredients = " ".join(df["text_ingredients"].dropna())
words = text_ingredients.split()
# remove punctuation
words = [word for word in words if word.isalnum()]
# remove stop words
words = [word for word in words if not nlp.vocab[word].is_stop]
# remove numbers
words = [word for word in words if not word.isnumeric()]

counter = Counter(words)

top_10 = counter.most_common(10)

plt.bar([word[0] for word in top_10], [word[1] for word in top_10])
plt.xticks(rotation=45)
plt.show()
```
